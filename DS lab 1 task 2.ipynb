```python
import pandas as pd

df = pd.read_csv("Au_nanoparticle_dataset.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

new_df_for_plot = df[["N_total","N_bulk","N_surface","R_avg"]]

plt.figure(figsize=(20, 4))
for i, col in enumerate(new_df_for_plot.columns):
    plt.subplot(1, 4, i + 1)
    sns.histplot(new_df_for_plot[col], bins=30, color='skyblue')
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df_for_plot, kind="scatter", diag_kind="hist", plot_kws={"alpha": 0.6})
plt.suptitle("Pairplot - Histograms and Scatter", y=1.02)
plt.show()
```

```python
g = sns.PairGrid(new_df_for_plot)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
plt.suptitle("Customized PairGrid", y=1.02)
plt.show()
```
